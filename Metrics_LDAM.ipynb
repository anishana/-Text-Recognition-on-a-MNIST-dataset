```python
def plot_loss(df, title, type):

  loss=[]
  plt.figure(figsize=(20,10))
  epochs = range(1,df[0].shape[0]+1)

  # print(df[1])
  colors=['r','g','b','c','m','y','k','violet','indigo','gold']
  for i in range(10):
    loss.append(df[i]['Average loss'])
  
  for i in range(0,9):
    plt.plot(epochs, loss[i], colors[i], label='Average loss '+type+' Epoch '+str(i))

  plt.title(title)
  plt.xlabel('Batch')
  plt.ylabel('Average Loss')
  plt.legend()
  plt.show()
```

```python
df = []
list_of_filenames = ['imbalanced','balanced','imbalanced_asym', 'balanced_asym','imbalanced_sym','balanced_sym']
for noise_type in list_of_filenames:
  for i in range(10):
    df.append(pd.read_csv('/content/drive/My Drive/CSE555/LDAM/'+noise_type+'/epoch_'+str(i)+'_'+noise_type+'_LDAM_training.csv', index_col=[0]))
  plot_loss(df,'Average Training Loss for LDAM for '+noise_type+' Dataset', 'Training')
  df=[]
```
